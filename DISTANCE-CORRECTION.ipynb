```python
# NEO4J database call for GROUP 1 - Close and Connected
graph = Graph()
c1 = '''MATCH (n:ROI)-[r:NOS]->(m:ROI)
WHERE r.weight > 2500 and r.length>1 and n.hemisphere="L"
RETURN n.name AS n1 ,m.name AS n2,r.length as D,r.weight as W,r._length,r._weight
ORDER BY r.length
limit 8'''

A = graph.run(c1)
# Plotting the data
s1 = []
for i,w in enumerate(A):
    a = w['r._length'] 
    b = w['r._weight']
    ### Pre-Filtering
    c = sorted(zip(a,b))
    d = [xx for xx in c if (xx[1]>200 and xx[0]>1)]
    x = [xx[0] for xx in d]
    y1 = [xx[1] for xx in d]
    if (len(x)<10):
        y = [num.log(xx) for xx in y1]
    else:
        _unused, y = correct(x,y1,10)

    p1 = figure(plot_width=500, plot_height=500, title='sample',y_axis_label = "Strength(log scale)",x_axis_label = "Distance")
#     p1.circle(a,[num.log(xx) for xx in b])
    p2 = figure(plot_width=500, plot_height=500, title='sample',y_axis_label = "Strength(log scale)",x_axis_label = "Distance")

    p1.square(x,[num.log(xx) for xx in y1],color='green',size=8)
    p2.circle(x,y,color='blue',size=8)
    s1.append([p1,p2])
grid = gridplot([[s1[0][0], s1[0][1]], [s1[1][0], s1[1][1]], [s1[2][0], s1[2][1]], [s1[3][0], s1[3][1]]])
show (grid)
```
