# SEED - ROI strength vs distance 
All the plots below are between ROIs that have at least a single seed connecting them by a strength larger than 2500 (Number of connected streamlines).  
The plots below are in three groups each containing 8 subplots.  
 - GROUP 1: ROIs are very close
 - GROUP 2: ROIs are very far 
 - GROUP 3: ROIs are very mid-range 

And the study is only for ROIS in right hemispher.

```python
from py2neo import Node, Relationship, Graph
from numpy.random import rand
import numpy as num
from bokeh.plotting import output_notebook,figure, show
from bokeh.layouts import row,column,gridplot
output_notebook()
```

```python
%%javascript
IPython.OutputArea.prototype._should_scroll = function(lines) {
    return false;
}
```

### GROUP 1: NEAR

```python
# NEO4J database call for GROUP 1
graph = Graph()
c1 = '''MATCH (n:ROI)-[r:NOS]->(m:ROI)
WHERE r.weight > 2500 and r.length>1
RETURN n.name AS n1 ,m.name AS n2,r.length as D,r.weight as W,r._length,r._weight
ORDER BY r.length
limit 8'''
A = graph.run(c1)
# Plotting the data
s = []
for i,w in enumerate(A):
    p = figure(plot_width=500, plot_height=500, title=w['n1']+'->'+w['n2'],y_axis_label = "Strength",x_axis_label = "Distance")
    a = w['r._length']
    b = w['r._weight']
    p.circle(a,b)
    s = s + [p]
grid = gridplot([[s[0], s[1]], [s[2], s[3]], [s[4], s[5]], [s[6], s[7]]])
show (grid)
```

### GROUP 2: FAR

```python
# NEO4J database call for GROUP 1
graph = Graph()
c1 = '''MATCH (n:ROI)-[r:NOS]->(m:ROI)
WHERE r.weight > 2500 and r.length>1
RETURN n.name AS n1 ,m.name AS n2,r.length as D,r.weight as W,r._length,r._weight
ORDER BY r.length DESC
limit 8'''
A = graph.run(c1)
# Plotting the data
s = []
for i,w in enumerate(A):
    p = figure(plot_width=500, plot_height=500, title=w['n1']+'->'+w['n2'],y_axis_label = "Strength",x_axis_label = "Distance")
    a = w['r._length']
    b = w['r._weight']
    p.circle(a,b)
    s = s + [p]
grid = gridplot([[s[0], s[1]], [s[2], s[3]], [s[4], s[5]], [s[6], s[7]]])
show (grid)
```

### GROUP 3: MIDRANGE

```python
# NEO4J database call for GROUP 1
graph = Graph()
c1 = '''MATCH (n:ROI)-[r:NOS]->(m:ROI)
WHERE r.weight > 2500 and r.length>20
RETURN n.name AS n1 ,m.name AS n2,r.length as D,r.weight as W,r._length,r._weight
ORDER BY r.length
limit 8'''
A = graph.run(c1)
# Plotting the data
s = []
for i,w in enumerate(A):
    p = figure(plot_width=500, plot_height=500, title=w['n1']+'->'+w['n2'],y_axis_label = "Strength",x_axis_label = "Distance")
    a = w['r._length']
    b = w['r._weight']
    p.circle(a,b)
    s = s + [p]
grid = gridplot([[s[0], s[1]], [s[2], s[3]], [s[4], s[5]], [s[6], s[7]]])
show (grid)
```
